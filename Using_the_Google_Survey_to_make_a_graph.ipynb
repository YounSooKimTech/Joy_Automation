```python
from google.colab import auth
auth.authenticate_user()

import gspread
from google.auth import default
creds, _ = default()

gc = gspread.authorize(creds)

worksheet = gc.open('MI_Qs').sheet1

# get_all_values gives a list of rows.
rows = worksheet.get_all_values()
print(rows)
```

```python
import pandas as pd
header = rows[0]
df = pd.DataFrame.from_records(rows[1:], columns=header)

df.iloc[:, 2:-1] = df.iloc[:, 2:-1].apply(pd.to_numeric, errors='coerce').astype('Int64')

df
```

```python
df = df.drop_duplicates(subset = ["Email Address"], keep='last')
#df = df.set_index("Email Address")

df
```

```python
import pandas as pd

temp = pd.DataFrame()
colnames = ["Verbal", "Logical", "Visual", "Interpersonal", "Musical", "Naturalistic", "Body", "Intrapersonal"
                                  ]
temp["email"] = df["Email Address"]
temp["name"] = df.iloc[:, -1]


# Repeat the process 3 times
for i in range(8):
    # Calculate the sum for each category based on 8 columns starting from column index 2
    start_col = 2 + i * 8
    end_col = start_col+1 + 7
    #print(start_col, end_col)
    col_name = colnames[i]
    #print(col_name)
    temp[col_name] = df.iloc[:, start_col:end_col].sum(axis=1)


temp
```

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

for index, row in temp.iterrows():
    email = row['email']
    name = row["name"]
    values = row[2:]

    palette = sns.color_palette("colorblind")

    # Create a horizontal bar graph
    plt.figure(figsize=(10, 6))
    ax = plt.gca()
    ax.bar(values.index, values, color=palette)
    plt.title(f'Bar Graph for {name}')
    plt.xlabel('Categories')
    plt.ylabel('Values')

    ax.axhline(y=8, color='b', linestyle='--', label='y=8')
    #ax.axhline(y=40, color='b', linestyle='--', label='y=40')

    # Save the graph with the email as the file name
    file_name = f"{email}.png"
    plt.savefig(file_name, bbox_inches="tight")
    plt.show()
    print("\n")
    plt.close()
```

```python
import pandas as pd

# Initialize an empty DataFrame to store the results
result_df = pd.DataFrame(columns=["Email Address", "verbal_sum", "logical_sum",
                                  "visual_sum", "interpersonal sum", "musical_sum",
                                  "naturalistic_sum", "bodyy_sum", "interpersonal_sum"
                                  ])

# Repeat the process 8 times
for i in range(3):
    # Calculate the sum for the verbal and math columns
    verbal_sums = df.iloc[:, 2 + i * 8:10 + i * 8].sum(axis=1)
    math_sums = df.iloc[:, 11 + i * 8:19 + i * 8].sum(axis=1)

    # Create a new DataFrame with the sums and an "id" column
    temp_df = pd.DataFrame({
        "Email Address": df["Email Address"],
        "verbal_sum": verbal_sums,
        "math_sum": math_sums
    })

    # Append the results to the main result DataFrame
    result_df = result_df.append(temp_df, ignore_index=True)

# The 'result_df' DataFrame now contains the sums for each set of columns with an "id" column.

result_df
```
