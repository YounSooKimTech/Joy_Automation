```python
# sample dataframe
# if you want to apply this code to your dataset,
# df = pd.read_csv("filepath/filename")

import pandas as pd

df = pd.DataFrame({"Text":["안녕하세요",
                           "Buenos Dias",
                           "Comment allez-vous",
                           "谢谢",
                           "Здравствуйте!"]})

df
```

# Language detection

```python
# https://spacy.io/usage/models

import spacy
from langdetect import detect

def detect_language(text):
    try:
        return detect(text)
    except:
        return "Unknown"

df['Lang'] = df['Text'].apply(detect_language)
df
```

```python
# detect the langugage
# https://en.wikipedia.org/wiki/List_of_ISO_639-1_codes

from langdetect import detect

def detect_language(text):
    try:
        return detect(text)
    except:
        return "unknown"

df["Lang"] = df["Text"].apply(detect_language)

df
```

# Check the lang type

```python
# this code print the text whose detected lang is not in major lang
# you can customize major_lang
# For convinience, I used three langs (English, Spanish, French)

lang_list = df["Lang"].value_counts().index.tolist()
major_lang = ["en", "es", "fr"]

for a_lang in lang_list:
  if a_lang not in major_lang:
    subset_df = df.loc[df["Lang"] == a_lang, ["Text", "Lang"]]
    print(subset_df, "\n")
```

# Translate the lang

```python
# translate the text into English

from deep_translator import GoogleTranslator

def translate_in_english(text):
  translated = GoogleTranslator(source="auto", target="en").translate(text)
  return translated

df["translated"] = df["Text"].apply(translate_in_english)

df
```
